```python
import matplotlib.pyplot as plt

ax = plt.subplot()

x = grafico['fabricante']
y = grafico['classificacao']

ax.bar(x, y)

plt.show()
```

```python
import matplotlib.pyplot as plt

ax = plt.subplot()

x = grafico['fabricante']
y = grafico['classificacao']

ax.bar(x, y)

ax.set_ylabel('Classificação')
ax.set_title('Gráfico de Ranking dos Fabricantes')

plt.savefig('grafico_rankinkg_fabricantes.png')

plt.show()
```
