```python
import pandas

linha_coluna={'A':['0A','1A','2A','3A','4A'],
            'B':['0B','1B','2B','3B','4B'],
            'C':['0C','1C','2C','3C','4C'],
            'D':['0D','1D','2D','3D','4D'],
            'E':['0E','1E','2E','3E','4E']}

dataframe=pandas.DataFrame(linha_coluna)
```

```python
dataframe
```

```python
dataframe[::]
```

```python
dataframe[2::]
```

```python
# ordem inversa
dataframe[::-1]
```

```python
from IPython.display import Image
Image('pandas-different-between-loc-vs-iloc.png')
```

```python
# por coluna 
print(dataframe.loc[:,'B':'D'])
```

```python
# por coluna
print(dataframe.loc[:,['B','A','D']])
```

```python
# por coluna
print(dataframe.loc[:,['B','C','D']])
```

```python
print(dataframe.loc[:,['C']])
```

```python
# por linha e coluna 
print(dataframe.loc[1,'D'])
```

```python
# por linha
print(dataframe.loc[2,:])
```

```python
from IPython.display import Image
Image('pandas-different-between-loc-vs-iloc.png')
```

```python
# por linha - fatiamento
print(dataframe.loc[3:])
```

```python
# por linha e coluna - fatiamento
print(dataframe.loc[3:,['C','D']])
```

```python
# por linha e coluna com step 
print(dataframe.loc[2::2,'B'::2])
```
